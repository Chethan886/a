```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('data.csv')

# Bar Chart: Average scores per subject
def plot_bar_chart():
    avg_scores = df[['Math', 'Science', 'English']].mean()
    plt.figure(figsize=(10,6))
    avg_scores.plot(kind='bar', color=['blue', 'green', 'orange'])
    plt.xlabel('Subjects')
    plt.ylabel('Average Score')
    plt.title('Average Scores per Subject')
    plt.show()

# Line Chart: Scores over students
def plot_line_chart():
    plt.figure(figsize=(10,6))
    for subject in ['Math', 'Science', 'English']:
        plt.plot(df['Name'], df[subject], marker='o', label=subject)
    plt.xlabel('Student')
    plt.ylabel('Score')
    plt.title('Scores of Students in Different Subjects')
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()

# Histogram: Distribution of Math scores
def plot_histogram():
    plt.figure(figsize=(10,6))
    plt.hist(df['Math'], bins=5, color='orange', edgecolor='black')
    plt.xlabel('Math Scores')
    plt.ylabel('Frequency')
    plt.title('Distribution of Math Scores')
    plt.show()

# Scatter Plot: Math vs Science scores
def plot_scatter_plot():
    plt.figure(figsize=(10,6))
    plt.scatter(df['Math'], df['Science'], color='red')
    plt.xlabel('Math Scores')
    plt.ylabel('Science Scores')
    plt.title('Scatter Plot of Math vs Science Scores')
    plt.show()

# Plot all charts
plot_bar_chart()
plot_line_chart()
plot_histogram()
plot_scatter_plot()
```
